```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_path ='/content/pexels-vika-glitter-392079-1648387.jpg' #here you have some work to do the thing is once you upload the photo you want in the files of colab copy the path and paste it
image = cv2.imread(image_path)
image
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5,minSize=(30,30))
```

```python
for(x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),6)
  roi_gray = gray_image[y:y+h,x:x+w]
  roi_color = image[y:y+h,x:x+w]
  eyes = eye_cascade.detectMultiScale(roi_gray)

  for (ex,ey,ew,eh) in eyes:
       cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),6)

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))

plt.title('FACE DETECTION')
plt.axis('off')
plt.show()
```
